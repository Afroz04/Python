 # EXAMPLE 1 : Basic Probability Calculation

```python
import random

#probability of an evnt
def probability (event_outcomes,total_outcomes):
    return event_outcomes / total_outcomes

#probability of rolling a 3 on a six-sided die 
event_outcomes = 1 
total_outcomes = 6 

prob = probability (event_outcomes,total_outcomes)
print (f"probability of rolling a 3: {prob}")
```

 # Example 2 : Simulating a Coin Toss

```python
import random

def coin_toss():
    return random.choice(['Heads', 'Tails'])

tosses = [coin_toss() for _ in range(1000)]
heads_count = tosses.count('Heads')
tails_count = tosses.count('Tails')

print(f"Heads: {heads_count}, Tails: {tails_count}")
```

 # Example 3 : Probability Distribution (Discrete)

```python
import matplotlib.pyplot as plt

outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6 

plt.bar(outcomes, probabilities)
plt.xlabel('Die outcome')
plt.ylabel('Probability')
plt.title('Probability Distribution of a Six-Sided Die')
plt.show()
```

 # Example 4 : Simulating Random Events (Monte Carlo Simulation)

```python
import random

def roll_dice():
    return random.randint(1, 6), random.randint(1, 6)

def simulate_dice_rolls(num_simulations):
    success_count = 0
    for _ in range(num_simulations):
        roll1, roll2 = roll_dice()
        if roll1 + roll2 == 7:
            success_count += 1
    return success_count / num_simulations

num_simulations = 10000
estimated_prob = simulate_dice_rolls(num_simulations)
print(f"Estimated Probability of rolling a sum of 7: {estimated_prob}")
```

# Example 5 : Continuous Probability Distribution (Normal Distribution)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


mu, sigma = 0, 1  
data = np.random.normal(mu, sigma, 1000)


plt.hist(data, bins=30, density=True, alpha=0.6, color='g')


xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Normal Distribution')
plt.show()
```
